DATA PREPROCESSING

```python
#Seperating Data For Training and Testing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x=df.drop(['smoking'],axis=1)
y=df['smoking']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=1)
```

```python
from sklearn.preprocessing import MinMaxScaler,RobustScaler,StandardScaler
```

```python
scaler=MinMaxScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
sc=StandardScaler()
sc
```

```python
x=sc.fit_transform(x)
x
```

```python
clf=LogisticRegression()
clf
```

```python
clf.fit(x_train,y_train)
```

```python
pred=clf.predict(x_test)
pred
```

```python
y_test
```

```python
clf.score(x_test,y_test)*100
```

```python
clf.predict([[1,1,40,155,60,81.3,1.2,1,1,1,114,73,94,215,82,73,126,12.9,1,0.7,18,19,27,1,0,1]])
```

```python
from sklearn.metrics import confusion_matrix
confmat = confusion_matrix(y_test, pred)

# Print the confusion matrix
print(confmat)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = [
    {'penalty' : ['l1', 'l2', 'elasticnet', 'none'],
    'C' : np.logspace(-4, 4, 20),
    'solver' : ['lbfgs','newton-cg','liblinear','sag','saga'],
    'max_iter' : [100, 1000,2500, 5000]
    }
]
```

```python
gs = GridSearchCV(clf, param_grid = param_grid, cv = 3, verbose=True, n_jobs=-1)
```

```python
best_clf=clf.fit(x,y)
```

```python
print("Accuracy - :",{best_clf.score(x,y)})
```
